Types of objects in python:
       1. Immutable Objects
       2. Mutuable Objects



  Immutable Objects:(whose value can't be changed once created)
    1. int
    2. float
    3. string
    4. bool
    5. tuple


  Mutuable Objects:
    1. List
    2. Set
    3. Dictionary

List

```python
# list should be invluded in the square brackets
my_list =[1,2,3,4,5]
print(my_list)
type(my_list)
```

```python
# lists can have multiple data types
my_list =[2,3,1.8,'English',True]
print(my_list)
```

```python
#Lists are mutable --> Changeable
#add elements to a list
my_list =[2,3,1.8,'English',True]
my_list.append(9)
print(my_list)


```

```python
#print elements of a list using their index
print(my_list[0])
print(my_list[2])
```

```python
#lists allowed duplicate values
list_1 = [1,2,3,4,5,8,91]
print(list_1)
```

```python
print(len(list_1))
```

```python
#initiate an empty list
list_3 = []
print(list_3)
```

```python
list_3.append(6)
print(list_3)
```

```python
#delete an item in a list
list_3 = [2,3,1.8,'English','True']
print(list_3)

del list_3[2]
print(list_3)
```

```python
#join two lists
list_4 = [1,2,3,4,5]
list_5 = [6,7,8,9,10]

list_6 =list_4 + list_5
print(list_6)
```

Tuple


```python
tuple_1 = (2,3,4,5)
print(tuple_1)
type(tuple_1)
```

```python
#tuple allows multiple data types
tuple_2 = (1,2,3.5,'Machine Learning',False)
print(tuple_2)
```

```python
#converting a list to tuple
my_list = [3,4,5,6]
print(my_list)

my_tuple = tuple(my_list)
print(my_tuple)
```

```python
print(my_tuple[0])
print(my_tuple[1])
```

Tuples are immutable -->Unchangeable

```python
print(len(my_tuple))
```

set

```python
#set --> curly brackets
my_set = {1,2,3,4,5}
print(my_set)
type(my_set)
```

```python
print(my_set[0])
```

```python
#convert a list to a set
list_5= [4,5,6,7,8]

x = set(list_5)
print(x)
```

```python
#set does not allowed duplicate values
set_3 = {1,2,3,4,5,6,7,6,4}
print(set_3)
```

Dictionary

Key _value Pair

```python
my_dictionary = {'name':'David','age':30 , 'country' : 'India'}
print(my_dictionary)
type(my_dictionary)

```

```python
print(my_dictionary['name'])
print(my_dictionary['age'])
print(my_dictionary['country'])
```

```python
# dictionary does not allow duplicate values
dictionary_2= {'name':'David', 'age' : 30,'country': 'india'}
print(dictionary_2)
```
